```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_spectrogram(S, m, sampling_rate):
    """
    Calculate the spectrogram of a signal.

    Parameters:
        S (array-like): The input signal of shape (N,).
        m (float): The interval width in seconds for splitting the signal.
        sampling_rate (int): The sampling rate of the signal (samples per second).

    Returns:
        F (2D array): The spectrogram represented as F.
    """
    # Step 1: Split the signal into intervals of width 'm' (in seconds)
    interval_width_samples = int(m * sampling_rate)
    num_intervals = len(S) // interval_width_samples
    intervals = np.array_split(S[:num_intervals * interval_width_samples], num_intervals)

    # Step 2: Apply Fast Fourier Transform (FFT) to each interval to generate frequency features
    fft_result = [np.fft.fft(interval) for interval in intervals] # 频率个数
    print(fft_result[0].shape, len(fft_result))

    # Step 3: Convert the FFT result to magnitude or power (spectrogram representation)
    # You can take the magnitude or power of the complex FFT result
    # depending on the desired representation.
    # Magnitude: `F = np.abs(fft_result)`
    # Power: `F = np.abs(fft_result) ** 2`
    F = np.abs(fft_result) 

    return fft_result, F

# Example usage:
# Assuming S is a 1D array representing the signal and sampling_rate is an integer.
# spectrogram = calculate_spectrogram(S, m=0.25, sampling_rate=your_sampling_rate)
```
